```python
import os

base_dir = r'C:\Users\larak\OneDrive\Documents\History-Lab\ddo\OCR paper\ddo'

# Initialize a list to store the lengths
lengths = []

# Iterate through the DataFrame
for file_name in df['file']:
    file_path = os.path.join(base_dir, file_name)
    try:
        with open(file_path, 'r', encoding='utf-8') as file:
            content = file.read()
            lengths.append(len(content))  # Append the length of the content
    except Exception as e:
        print(f"Failed to read {file_path}: {e}")

# Create a DataFrame from lengths
length_df = pd.DataFrame(lengths, columns=['Length'])

# Plotting the distribution of document lengths
plt.figure(figsize=(10, 6))
plt.hist(length_df['Length'], bins=30, color='blue', alpha=0.7)
plt.title('Distribution of Document Lengths')
plt.xlabel('Document Length')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```
